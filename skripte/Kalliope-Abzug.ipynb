# Import der Metadaten zur Göschen-Sammlung aus Kalliope

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup as soup
import xmltodict
from datetime import datetime
import re
```

```python
def sru():
    
    base_url = "https://kalliope-verbund.info/sru"
    params = {'recordSchema' : 'mods37',
          'operation': 'searchRetrieve',
          'version': '1.2',
          'maximumRecords': '100',
          'query': "ead.repository.isil=DE-101d",
         }
    r = requests.get(base_url, params=params)
    xml = soup(r.content, features="xml")
    records = xml.find_all('srw:record')
    
    if len(records) < 100:
        # print(r.url)
        return records
    
    else:
        
        num_results = 100
        i = 101
        while num_results == 100:
            
            params.update({'startRecord': i})
            r = requests.get(base_url, params=params)
            xml = soup(r.content, features="xml")
            new_records = xml.find_all('srw:record')
            records+=new_records
            i+=100
            num_results = len(new_records)
            
        return records
```

```python
daten = list()
for record in sru():
    # print(xmltodict.parse(record.find('srw_dc:dc').prettify())["srw_dc:dc"])
    
    #try:
        #daten.append(xmltodict.parse(record.find('srw_dc:dc').prettify())["srw_dc:dc"])
    #except:
        #pass
    try:
        daten.append(xmltodict.parse(record.find('mods').prettify())["mods"])
    except:
        pass
        
    # for feld in record.find('srw_dc:dc'):
        # print(feld)

df = pd.DataFrame.from_dict(daten)
df
```

```python
# Die alten SWD-PPNs werden in aktuelle GND-PPNs umgerechnet, betrifft vor allem prüfziffer

def ppn_check(ppn):
    if "-" in ppn:
        counter = 2
        summe = 0
        for ziffer in ppn[:-2][::-1]:
            summe += int(ziffer) * counter
            counter += 1
        pruefziffer = (11 - (summe % 11) ) % 11
        pruefziffer = "X" if pruefziffer == 10 else pruefziffer
          
        return (8 - len(ppn[:-2])) * "0" + ppn[:-2] + str(pruefziffer)
    else:
        return ppn
```

```python
role_replace = {
    "cre":"aut",
    "rcp":"rcp",
    "edt":"edt"}

role_text = {
    "cre":"Verfasser",
    "rcp":"Adressat",
    "edt":"Herausgeber"}

find_tq = {
    "a":"1032507942",
    "b":"1032507977",
    "c":"1032507993",
    "d":"1032508027",
    "g":"103250806X"}

def datensatz_schreiben(row):
    # print(row["originInfo"])
    # print(row['identifier']['#text'])
    pica = """0500 Da
0600 yy\n"""
    

    if type(row["originInfo"]["dateCreated"]) == list:
        # pica += f'1100 {row["originInfo"]["dateCreated"][0]["#text"][:4]}\n1110 {row["originInfo"]["dateCreated"][-1]}'

        if year := re.search(r"\d{4}", row["originInfo"]["dateCreated"][0]["#text"]):
            pica += f"1100 {year[0]}"
            if len(year.string) > 4:
                pica += f"$n{row['originInfo']['dateCreated'][-1]}"
            
            pica += f"\n1110 {row['originInfo']['dateCreated'][-1]}"

        if len(row["originInfo"]["dateCreated"][0]["#text"]) == 10:
            pica += f'*{row["originInfo"]["dateCreated"][0]["#text"][8:10]}.{row["originInfo"]["dateCreated"][0]["#text"][5:7]}.{row["originInfo"]["dateCreated"][0]["#text"][:4]}'
        pica += '$4ezth\n'

    elif type(row["originInfo"]["dateCreated"]) == str:
        if year := re.search(r"\d{4}",row["originInfo"]["dateCreated"]):
            pica += f"1100 {year[0]}$n{year.string}\n1110 {year.string}$4ezth\n"
        else:
            pica += f"1100 1900$n[nicht nach 1900]\n1110 {row['originInfo']['dateCreated']}\n"
            # alle Datensätze, die so mangelhafte Datumsinformationen haben, werden in eine Liste geschrieben, damit CS diese nachrecherchieren kann.
            with open("../dat/od-fehlerliste.txt", 'a') as file:
                file.write(f"{row['location']['shelfLocator']}, {row['originInfo']['dateCreated']}\n")
    pica += f"""1130 TB-papier
1130 !040445224!
1131 !040082407!
1132 a1-analog;a2-hand;f1-text;f2-blatt;e2-uv;o-org;v-ganz
1500 /1{row['language']['languageTerm'][0]['#text']}
"""
    personcount = 0
    if type(row['name']) == list:
        for person in row['name']:
            # gibt es eine GND-Verknüpfung für personenname?
            # wenn als GND-Verknüpfung unbekannt angegeben ist, wird die person nicht übernommen
            if (person['@authority'] == 'DE-588') and (person['namePart'].lower() != 'unbekannt'):
                pica += f"30{personcount}0 !{ppn_check(person['@valueURI'].split('/')[-1])}!"
                personcount += 1
            elif (person['namePart'].lower() != 'unbekannt'):
                pica += f"30{personcount}0 {person['namePart']}"
                personcount += 1

            if (type(person['role']['roleTerm']) == list) and (person['namePart'].lower() != 'unbekannt'):
                pica += f"$B{role_text[person['role']['roleTerm'][0]['#text']]}$4{role_replace[person['role']['roleTerm'][0]['#text']]}\n"
            
            elif (type(person['role']['roleTerm']) == dict) and (person['namePart'].lower() != 'unbekannt'):
                if person['role']['roleTerm']['#text'] == "Schreiber":
                    pass
                elif person['role']['roleTerm']['#text'] == "Korrespondenzpartner":
                    pica += f"$BKorrespondenzpartner$4krp\n"
                else:
                    pica += f"$B{person['role']['roleTerm']['#text']}$4oth\n"
            
    
    elif type(row['name']) == dict:
        if row['name']['@authority'] == 'DE-588':
            pica += f"3000 !{ppn_check(row['name']['@valueURI'].split('/')[-1])}!$B{role_text[row['name']['role']['roleTerm'][0]['#text']]}$4{role_replace[row['name']['role']['roleTerm'][0]['#text']]}\n"
        else:
            pica += f"3000 {row['name']['namePart']}$B{role_text[row['name']['role']['roleTerm'][0]['#text']]}$4{role_replace[row['name']['role']['roleTerm'][0]['#text']]}\n"
    
    pica += f"""4000 {row['titleInfo']['title']}
"""
    if row['originInfo']['place']['placeTerm']['#text'].lower() == "o.o.":
        pica += "4046 [Entstehungsort nicht ermittelbar]\n"
    else:
        pica += f"4046 {row['originInfo']['place']['placeTerm']['#text']}\n"
    
    pica += f"""4060 {row['physicalDescription']['extent']}
4105 !1059146037!
4105 !1077661932!
"""
    if type(row['note']) == dict:
        if row['note']['@type'] == 'encounteredVariantNames':
            pica += f"4201 alternative Namensform {row['note']['#text']}\n"
    elif type(row['note']) == list:
        for note in row['note']:
            if type(note) == dict:
                if note['@type'] == 'encounteredVariantNames':
                    pica += f"4201 alternative Namensform {note['#text']}\n"
    
    pica += "4700 |BSM|\n"

    if type(row['abstract']) != float:
        pica += f"4705 {row['abstract']['#text']}\n"
    
    pica += """5560 [RNA]Korrespondenz
5590 [Objektgattung]
5590 !040082407!
"""
    if len(row['originInfo']['place']['placeTerm']) > 2:
        pica += f"5591 [Entstehungsort]\n5591 !{ppn_check(row['originInfo']['place']['placeTerm']['@valueURI'].split('/')[-1])}!\n"
    pica += f"""7001 {now.strftime('%d-%m-%y')} : x
4800 !009033645!
6710 !{find_tq[row['location']['shelfLocator'].split('/',1)[1][:1].lower()]}!
7100 {row['location']['shelfLocator']} @ m
7109 !!DBSM/S!! ; {row['location']['shelfLocator']}"""
    pica += "\n\t\n"
    
    return pica
```
