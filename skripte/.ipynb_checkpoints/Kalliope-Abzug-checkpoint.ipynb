# Import der Metadaten zur Göschen-Sammlung aus Kalliope

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup as soup
import xmltodict
from datetime import datetime
```

```python
def sru():
    
    base_url = "https://kalliope-verbund.info/sru"
    params = {'recordSchema' : 'mods37',
          'operation': 'searchRetrieve',
          'version': '1.2',
          'maximumRecords': '100',
          'query': "ead.repository.isil=DE-101d",
         }
    r = requests.get(base_url, params=params)
    xml = soup(r.content, features="xml")
    records = xml.find_all('srw:record')
    
    if len(records) < 100:
        # print(r.url)
        return records
    
    else:
        
        num_results = 100
        i = 101
        while num_results == 100:
            
            params.update({'startRecord': i})
            r = requests.get(base_url, params=params)
            xml = soup(r.content, features="xml")
            new_records = xml.find_all('srw:record')
            records+=new_records
            i+=100
            num_results = len(new_records)
            
        return records
```

```python
daten = list()
for record in sru():
    # print(xmltodict.parse(record.find('srw_dc:dc').prettify())["srw_dc:dc"])
    
    #try:
        #daten.append(xmltodict.parse(record.find('srw_dc:dc').prettify())["srw_dc:dc"])
    #except:
        #pass
    try:
        daten.append(xmltodict.parse(record.find('mods').prettify())["mods"])
    except:
        pass
        
    # for feld in record.find('srw_dc:dc'):
        # print(feld)

df = pd.DataFrame.from_dict(daten)
df
```

# Umwandlung in Pica

```python
# Die alten SWD-PPNs werden in aktuelle GND-PPNs umgerechnet, betrifft vor allem prüfziffer

def ppn_check(ppn):
    if "-" in ppn:
        counter = 2
        summe = 0
        for ziffer in ppn[:-2][::-1]:
            summe += int(ziffer) * counter
            counter += 1
        pruefziffer = (11 - (summe % 11) ) % 11
        pruefziffer = "X" if pruefziffer == 10 else pruefziffer
          
        return (8 - len(ppn[:-2])) * "0" + ppn[:-2] + str(pruefziffer)
    else:
        return ppn
```

```python
role_replace = {
    "cre":"aut",
    "rcp":"rcp",
    "edt":"edt"}

role_text = {
    "cre":"Verfasser",
    "rcp":"Adressat",
    "edt":"Herausgeber"}

def datensatz_schreiben(row):
    # print(row["originInfo"])
    print(row['identifier']['#text'])
    pica = """0500 Da
0600 yy\n"""
    pica += f'1100 {row["originInfo"]["dateCreated"][-1][6:]}$n{row["originInfo"]["dateCreated"][-1]}\n1110 *{row["originInfo"]["dateCreated"][-1]}$4ezth\n'
    pica += f"""1130 TB-papier
1130 !040445224!Papier [Ts1]
1131 !040082407!Brief [Ts1]
1132 a1-analog;a2-hand;f1-text;f2-blatt;e2-uv;o-org;v-ganz
1500 /1{row['language']['languageTerm'][0]['#text']}
"""
    personcount = 0
    if type(row['name']) == list:
        for person in row['name']:
            if type(person['role']['roleTerm']) == list:
                pica += f"30{personcount}0 {person['@valueURI'].split('/')[-1]}$B{role_text[person['role']['roleTerm'][0]['#text']]}$4{role_replace[person['role']['roleTerm'][0]['#text']]}\n"
            elif type(person['role']['roleTerm']) == dict:
                pica += f"30{personcount}0 {person['@valueURI'].split('/')[-1]}$B{person['role']['roleTerm']['#text']}$4oth\n"
            personcount += 1
    
    elif type(row['name']) == dict:
        pica += f"3000 !{row['name']['@valueURI'].split('/')[-1]}!$B{role_text[row['name']['role']['roleTerm'][0]['#text']]}$4{role_replace[row['name']['role']['roleTerm'][0]['#text']]}\n"
    pica += f"""4000 {row['titleInfo']['title']}
"""
    if row['originInfo']['place']['placeTerm']['#text'].lower() == "o.o.":
        pica += "4046 [Entstehungsort nicht ermittelbar]\n"
    else:
        pica += f"4046 {row['originInfo']['place']['placeTerm']['#text']}\n"
    
    pica += f"""4060 {row['physicalDescription']['extent']}
4105 !1059146037!Archivalien und Dokumente zur Buchgeschichte
4105 !1077661932!Archivalien der Bibliothek des Börsenvereins der Deutschen Buchhändler zu Leipzig
4201 #todo
4700 |BSM|
4705 Regest
5560 [RNA]Korrespondenz
5590 [Objektgattung]
5590 !040082407!
"""
    if len(row['originInfo']['place']['placeTerm']) > 2:
        pica += f"5591 [Entstehungsort]\n5591 !{ppn_check(row['originInfo']['place']['placeTerm']['@valueURI'].split('/')[-1])}!\n"
    pica += "\n\t\n"
    
    return pica
```

```python
for index, row in df.sample(5).iterrows():
    print(datensatz_schreiben(row))
```

```python
now = datetime.now()
with open(f"../dat/goeschen-{now.strftime('%y-%m-%d-%H-%M-%S')}.dat","w") as f:
    with open(f"../dat/goeschen-recent.dat", 'w') as f2:
        for index, row in df.iterrows():
            f.write(datensatz_schreiben(row))
            f2.write(datensatz_schreiben(row))

with open(f"../dat/goeschen-recent-sample.dat", 'w') as f:
    for index, row in df.sample(5).iterrows():
        f.write(datensatz_schreiben(row))
```
